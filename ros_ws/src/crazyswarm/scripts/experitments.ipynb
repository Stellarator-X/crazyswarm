```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("testWaypoints.csv")
df.head()
```

```python
df.columns
```

```python
xs = [float(x) for x in df['x']]
ys = [float(x) for x in df['y']]
zs = [float(x) for x in df['z']]
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(xs, ys, zs)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
